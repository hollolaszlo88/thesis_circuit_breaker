## Section 3B — Per-Feature Raw Attribution Score Analysis

This section aggregates attribution at the feature level across questions.
For each feature `(layer, position, feature_idx)`, we average the **raw attribution score**
(`score`) over only the questions where that feature appears in the selected top-feature list.

So if a feature appears in 3 questions, its mean raw score is computed over those 3 rows only.

```python
# Aggregate raw attribution scores per recurring feature across observed questions only.
required_cols = {'qid', 'layer', 'position', 'feature_idx', 'score'}
missing = required_cols.difference(df_features.columns)
if missing:
    raise ValueError(f'df_features is missing required columns: {sorted(missing)}')

feat_scores = df_features.copy()
feat_scores['feature_key'] = list(
    zip(
        feat_scores['layer'].astype(int),
        feat_scores['position'].astype(int),
        feat_scores['feature_idx'].astype(int),
    )
)

feature_score_summary = (
    feat_scores
    .groupby('feature_key')
    .agg(
        n_questions=('qid', 'nunique'),
        n_rows=('qid', 'size'),
        mean_raw_score=('score', 'mean'),
        mean_abs_raw_score=('score', lambda s: s.abs().mean()),
        min_raw_score=('score', 'min'),
        max_raw_score=('score', 'max'),
    )
    .reset_index()
)

# Add per-feature question IDs for interpretability
qids_by_feature = (
    feat_scores
    .groupby('feature_key')['qid']
    .apply(lambda s: ', '.join(sorted(set(s))))
    .reset_index(name='qids_present')
)
feature_score_summary = feature_score_summary.merge(qids_by_feature, on='feature_key', how='left')

feature_score_summary = feature_score_summary.sort_values(
    ['mean_abs_raw_score', 'n_questions'], ascending=[False, False]
).reset_index(drop=True)

print('=== PER-FEATURE RAW SCORE SUMMARY (observed questions only) ===')
print(f'Unique features analysed: {len(feature_score_summary)}')
print('Columns: feature_key, n_questions, mean_raw_score, mean_abs_raw_score, min/max, qids_present')
print()
print(feature_score_summary.to_string(index=False))

# Focus view: features that recur in 3+ questions
recurring3 = feature_score_summary[feature_score_summary['n_questions'] >= 3].copy()
print('\n=== FEATURES APPEARING IN >=3 QUESTIONS ===')
if len(recurring3):
    print(recurring3[['feature_key', 'n_questions', 'mean_raw_score', 'mean_abs_raw_score', 'qids_present']].to_string(index=False))
else:
    print('None')

# Plot: top features by absolute mean raw score
plot_df = feature_score_summary.head(20).copy()
plot_df['feature_key_str'] = plot_df['feature_key'].astype(str)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Left: signed mean raw score
axes[0].barh(plot_df['feature_key_str'][::-1], plot_df['mean_raw_score'][::-1], color='slateblue', alpha=0.85)
axes[0].axvline(0, color='black', linewidth=0.8)
axes[0].set_xlabel('Mean raw score (score)')
axes[0].set_title('Top 20 features by |mean raw score|')

# Right: recurrence vs magnitude
axes[1].scatter(
    feature_score_summary['n_questions'],
    feature_score_summary['mean_abs_raw_score'],
    color='darkorange',
    alpha=0.8,
)
axes[1].set_xlabel('Number of questions feature appears in')
axes[1].set_ylabel('Mean absolute raw score')
axes[1].set_title('Feature recurrence vs attribution magnitude')

plt.tight_layout()
plt.savefig(RUN_DIR / 'tables' / 'feature_raw_score_summary.png', dpi=150, bbox_inches='tight')
plt.show()
print('\nFigure saved to', RUN_DIR / 'tables' / 'feature_raw_score_summary.png')

# Keep this table available for later conclusion cells
df_feature_score_summary = feature_score_summary
```

## Section 4 — Intervention Analysis

Circuit interventions test whether the recurring features are causally responsible for the
model's True/False decision. We examine:

- **Amplify (2x and 5x)**: does moderate vs strong amplification move margins in the expected direction?
- **Ablate (0x)**: does removing feature activity reduce margin or flip predictions?

Comparing 2x vs 5x helps identify nonlinear effects (e.g., useful features that become harmful
when over-amplified).

```python
# Cell A — Group summary (aligned vs opposed), including correctness-aware movement
intervention_types = ['amplify_2x', 'amplify_5x', 'ablate']

group_df = df_interv[df_interv['scope'] == 'group'].copy()
group_df['label_bool'] = group_df['label'].astype(str).str.lower().map({'true': True, 'false': False})
group_df['label_sign'] = group_df['label_bool'].map({True: 1.0, False: -1.0})
group_df['correctness_delta'] = group_df['delta_margin'] * group_df['label_sign']
group_df['moved_correct_direction'] = group_df['correctness_delta'] > 0

print('=== GROUP INTERVENTION SUMMARY ===')
print(f"{'Key':<36} {'avg_delta':>10}  {'avg_correct':>12}  {'pct_correct_move':>16}  {'n_flipped':>10}  {'total':>6}")
print('-' * 104)
for direction in ['aligned', 'opposed']:
    for itype in intervention_types:
        key = f'group_{direction}_{itype}'
        subset = group_df[(group_df['direction'] == direction) & (group_df['intervention'] == itype)]
        if subset.empty:
            continue
        avg_delta = subset['delta_margin'].mean()
        avg_correct = subset['correctness_delta'].mean()
        pct_correct_move = 100.0 * subset['moved_correct_direction'].mean()
        n_flipped = int(subset['flipped'].sum())
        total = len(subset)
        print(f"  {key:<34} {avg_delta:>+10.4f}  {avg_correct:>+12.4f}  {pct_correct_move:>15.1f}%  {n_flipped:>10}  {total:>6}")

print()
print('=== GROUP PER-QUESTION DETAILS ===')
cols = [
    'id', 'label', 'intervention', 'direction',
    'margin_before', 'margin_after', 'delta_margin', 'correctness_delta',
    'moved_correct_direction', 'flipped'
]
print(group_df[cols].to_string(index=False))
```

```python
# Cell B — Single-feature summary, including correctness-aware movement
single_df = df_interv[df_interv['scope'] == 'single'].copy()
single_df['label_bool'] = single_df['label'].astype(str).str.lower().map({'true': True, 'false': False})
single_df['label_sign'] = single_df['label_bool'].map({True: 1.0, False: -1.0})
single_df['correctness_delta'] = single_df['delta_margin'] * single_df['label_sign']
single_df['moved_correct_direction'] = single_df['correctness_delta'] > 0

if single_df.empty:
    print('No single-feature intervention records found.')
else:
    print('=== SINGLE-FEATURE INTERVENTION SUMMARY ===')
    agg = (single_df
           .groupby(['feature_key', 'intervention'])
           .agg(
               avg_delta=('delta_margin', 'mean'),
               avg_correct=('correctness_delta', 'mean'),
               pct_correct_move=('moved_correct_direction', lambda s: 100.0 * s.mean()),
               n=('delta_margin', 'count')
           )
           .reset_index())
    print(agg.to_string(index=False))

    print()
    print('=== SINGLE-FEATURE PER-QUESTION DETAILS ===')
    cols = [
        'id', 'label', 'feature_key', 'intervention',
        'margin_before', 'margin_after', 'delta_margin', 'correctness_delta',
        'moved_correct_direction', 'flipped'
    ]
    print(single_df[cols].sort_values(['feature_key', 'intervention']).to_string(index=False))
```

```python
# Cell C — Two-panel intervention plot (correctness-aware metric)
group_df  = df_interv[df_interv['scope'] == 'group'].copy()
single_df = df_interv[df_interv['scope'] == 'single'].copy()
intervention_types = ['amplify_2x', 'amplify_5x', 'ablate']
colors = {'amplify_2x': 'mediumpurple', 'amplify_5x': 'steelblue', 'ablate': 'tomato'}

for _df in [group_df, single_df]:
    _df['label_bool'] = _df['label'].astype(str).str.lower().map({'true': True, 'false': False})
    _df['label_sign'] = _df['label_bool'].map({True: 1.0, False: -1.0})
    _df['correctness_delta'] = _df['delta_margin'] * _df['label_sign']

fig, axes = plt.subplots(1, 2, figsize=(17, 5))

# ── Left panel: group aligned vs opposed (avg correctness_delta) ──────────
ax = axes[0]
labels_plot = ['aligned', 'opposed']
x = list(range(len(labels_plot)))
width = 0.25

for i, itype in enumerate(intervention_types):
    vals = []
    for direction in labels_plot:
        subset = group_df[(group_df['direction'] == direction) & (group_df['intervention'] == itype)]
        vals.append(subset['correctness_delta'].mean() if not subset.empty else 0.0)
    offsets = [xi + (i - 1) * width for xi in x]
    ax.bar(offsets, vals, width, label=itype, color=colors.get(itype), alpha=0.85)

ax.axhline(0, color='black', linewidth=0.8)
ax.set_xticks(x)
ax.set_xticklabels(labels_plot)
ax.set_ylabel('Avg correctness-adjusted delta')
ax.set_title('Group interventions: aligned vs opposed')
ax.legend()

# ── Right panel: single-feature correctness-adjusted deltas ───────────────
ax = axes[1]
if not single_df.empty:
    feat_keys = list(single_df['feature_key'].unique())
    x2 = list(range(len(feat_keys)))

    for i, itype in enumerate(intervention_types):
        vals = []
        for fk in feat_keys:
            sub = single_df[(single_df['feature_key'] == fk) & (single_df['intervention'] == itype)]
            vals.append(sub['correctness_delta'].mean() if not sub.empty else 0.0)
        offsets = [xi + (i - 1) * width for xi in x2]
        ax.bar(offsets, vals, width, label=itype, color=colors.get(itype), alpha=0.85)

    ax.axhline(0, color='black', linewidth=0.8)
    ax.set_xticks(x2)
    ax.set_xticklabels([str(fk) for fk in feat_keys], rotation=20, ha='right', fontsize=8)
    ax.set_ylabel('Avg correctness-adjusted delta')
    ax.set_title('Single-feature interventions (top aligned)')
    ax.legend()
else:
    ax.text(0.5, 0.5, 'No single-feature data', ha='center', va='center', transform=ax.transAxes)

plt.tight_layout()
out_path = RUN_DIR / 'tables' / 'intervention_deltas.png'
plt.savefig(out_path, dpi=150, bbox_inches='tight')
plt.show()
print(f'Figure saved to {out_path}')
```

## Section 4B — Single-Question Attribution Graph

Visualise one selected question's attribution graph directly in the notebook from
`graphs_pt/<qid>_<ACTIVE_TEMPLATE>.pt`.

```python
import sys
from pathlib import Path

# adjust if your repo is elsewhere
REPO_ROOT = Path("/Users/Lacus/Documents/ELTE/Thesis/circuit-tracer-main")
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))
```

```python
# Plot one selected question's attribution graph inline
# Change this to any question ID present in the run, e.g. 'TI-03'.
QID_TO_PLOT = 'TI-01'
NODE_THRESHOLD = 0.80
EDGE_THRESHOLD = 0.98
MAX_DRAW_EDGES = 250
plot_graph = False   # Only plot graph if True

if plot_graph:
    import os
    import sys
    import torch
    from pathlib import Path

    # Mirror the robust repo discovery used in 1_run_attribution.ipynb.
    def _find_repo_root() -> Path | None:
        start = Path.cwd().resolve()
        for directory in [start, *start.parents]:
            if (directory / 'circuit_tracer' / '__init__.py').is_file():
                return directory

        workspace = Path('/workspace')
        if workspace.is_dir():
            for child in workspace.iterdir():
                if child.is_dir() and (child / 'circuit_tracer' / '__init__.py').is_file():
                    return child

        repo_override = os.environ.get('CT_REPO_DIR')
        if repo_override:
            override_path = Path(repo_override).expanduser().resolve()
            if (override_path / 'circuit_tracer' / '__init__.py').is_file():
                return override_path

        return None

    _repo_root = _find_repo_root()
    if _repo_root is None:
        raise ModuleNotFoundError(
            'Could not locate repo root containing `circuit_tracer/__init__.py`. '
            'Set CT_REPO_DIR to your repo path or open notebook from repo root.'
        )

    if str(_repo_root) not in sys.path:
        sys.path.insert(0, str(_repo_root))

    print(f'Repo root on sys.path: {_repo_root}')
    from circuit_tracer.graph import Graph, prune_graph

    try:
        import networkx as nx
    except ImportError:
        nx = None

    pt_path = RUN_DIR / 'graphs_pt' / f'{QID_TO_PLOT}_{ACTIVE_TEMPLATE}.pt'
    if not pt_path.is_file():
        raise FileNotFoundError(f'Graph file not found: {pt_path}')

    graph = Graph.from_pt(str(pt_path), map_location='cpu')
    node_mask, edge_mask, _ = prune_graph(graph, node_threshold=NODE_THRESHOLD, edge_threshold=EDGE_THRESHOLD)

    n_features = len(graph.selected_features)
    n_tokens = len(graph.input_tokens)
    n_logits = len(graph.logit_targets)
    error_start = n_features
    error_end = error_start + n_tokens * graph.cfg.n_layers
    token_start = error_end
    token_end = token_start + n_tokens

    labels = {}
    for idx in node_mask.nonzero(as_tuple=False).squeeze(-1).tolist():
        if idx < n_features:
            layer, pos, feat_idx = graph.active_features[graph.selected_features[idx]].tolist()
            labels[idx] = f'F[L{layer},p{pos},f{feat_idx}]'
        elif idx < error_end:
            layer, pos = divmod(idx - error_start, n_tokens)
            labels[idx] = f'E[L{layer},p{pos}]'
        elif idx < token_end:
            pos = idx - token_start
            labels[idx] = f'T[{pos}]'
        else:
            logit_pos = idx - token_end
            tok, vocab_idx = graph.logit_targets[logit_pos]
            labels[idx] = f'Logit[{tok or vocab_idx}]'

    edge_rows, edge_cols = edge_mask.nonzero(as_tuple=True)
    weights = graph.adjacency_matrix[edge_rows, edge_cols]

    # Keep only edges whose endpoints survive node pruning
    keep = []
    active_nodes = set(node_mask.nonzero(as_tuple=False).squeeze(-1).tolist())
    for i, (dst, src) in enumerate(zip(edge_rows.tolist(), edge_cols.tolist())):
        if src in active_nodes and dst in active_nodes:
            keep.append(i)

    if not keep:
        print('No edges survived pruning for this threshold pair. Try lower thresholds.')
    else:
        # Sort by absolute edge weight for cleaner display
        keep = sorted(keep, key=lambda i: float(abs(weights[i])), reverse=True)[:MAX_DRAW_EDGES]

        if nx is None:
            print('networkx is not installed; showing top edges table instead.')
            rows = []
            for i in keep[:30]:
                dst, src = int(edge_rows[i]), int(edge_cols[i])
                rows.append({
                    'source': labels.get(src, str(src)),
                    'target': labels.get(dst, str(dst)),
                    'weight': float(weights[i]),
                })
            pd.DataFrame(rows)
        else:
            G = nx.DiGraph()
            for i in keep:
                dst, src = int(edge_rows[i]), int(edge_cols[i])
                G.add_edge(labels.get(src, str(src)), labels.get(dst, str(dst)), weight=float(weights[i]))

            # Layout + style
            pos = nx.spring_layout(G, seed=7, k=0.9 / (len(G.nodes()) ** 0.5 if len(G.nodes()) > 1 else 1.0))
            edge_w = [abs(G[u][v]['weight']) for u, v in G.edges()]
            max_w = max(edge_w) if edge_w else 1.0
            edge_w_scaled = [0.5 + 3.0 * (w / max_w) for w in edge_w]

            plt.figure(figsize=(16, 10))
            nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightsteelblue', edgecolors='black', linewidths=0.5)
            nx.draw_networkx_edges(G, pos, width=edge_w_scaled, arrows=True, arrowsize=10, alpha=0.7)
            nx.draw_networkx_labels(G, pos, font_size=8)
            plt.title(
                f'Attribution Graph — {QID_TO_PLOT} ({ACTIVE_TEMPLATE})\n'
                f'nodes={G.number_of_nodes()} edges={G.number_of_edges()} '
                f'node_thr={NODE_THRESHOLD} edge_thr={EDGE_THRESHOLD}'
            )
            plt.axis('off')
            plt.tight_layout()
            plt.show()

            print(f'Loaded: {pt_path.name}')
            print(f'Pruned graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges (drawn max {MAX_DRAW_EDGES})')
```

## Section 5 — Conclusions (Triangle Inequality)

### 1) Baseline behavior

Across the active template, the model predicted **`True` for every question**.
This is visible in the baseline margins being positive throughout.
The practical consequence is that all False-labelled items are errors of the same type
(model over-accepts triangle validity), indicating a strong response bias toward `True`.

### 2) What attribution shows

The attribution tables identify recurring internal features that are reused across multiple
triangle-inequality prompts. This supports a **shared circuit hypothesis**: the model is not
solving each question independently, but reusing a stable internal mechanism.

### 3) What interventions show causally

The sign-aware intervention analysis gives two main findings:

- Group-level interventions produce only small average margin shifts and no label flips.
  This means the selected recurring groups influence confidence, but do not dominate the final decision.
- Single-feature interventions are more diagnostic. For the strongest recurring features
  (e.g. `(0, 2, 16200)` and `(1, 1, 11660)`), both ablation and 5x amplification often reduce the
  True-False margin. This pattern is consistent with **nonlinear / regime-limited behavior**:
  the feature is useful near its natural activation range, while extreme scaling can destabilize
  downstream computations.

### 4) Overall interpretation

The model appears to use a coherent, recurring internal circuit for triangle-inequality reasoning,
but this circuit currently supports a decision policy that is biased toward predicting `True`.
In other words, the model has reusable internal structure, yet its final classification boundary in
this task setting is miscalibrated.

### 5) Implications for thesis narrative

This run is a good example of how attribution plus intervention can separate:

- **mechanistic reuse** (recurring features are real), from
- **task-level correctness** (outputs are systematically biased).

That distinction is valuable: a model can be internally consistent and still wrong in a
systematic way.
